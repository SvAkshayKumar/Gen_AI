```python
from groq import Groq
import os
import time
from google.colab import userdata


# Initialize Groq client
client = Groq(api_key=userdata.get('GROQ_API'))

def ask_ai(prompt):
    time.sleep(0.5)  # light rate-limit safety
    response = client.chat.completions.create(
        model="llama-3.1-8b-instant",
        messages=[
            {"role": "user", "content": prompt}
        ],
        temperature=0.7
    )
    return response.choices[0].message.content
```

Exercise 3
Use of Persona Prompt, Root Prompt, Cognitive Verifier Pattern, Question Refinement Pattern, Provide New Information and Ask Questions

```python
print("\n=== EXERCISE 3 ===\n")

# 1. Root Prompt
root_prompt = """
You are an expert Generative AI tutor.
Explain concepts clearly, step by step, using simple language.
"""

print("Root Prompt Output:\n")
print(ask_ai(root_prompt + "Explain Prompt Engineering."))
print("-" * 80)

# 2. Persona Prompt
persona_prompt = """
Act as a college professor teaching Generative AI to undergraduate students.
Explain Prompt Engineering with real-life examples.
"""

print("Persona Prompt Output:\n")
print(ask_ai(persona_prompt))
print("-" * 80)

# 3. Cognitive Verifier Pattern
cognitive_verifier_prompt = """
Explain what Prompt Engineering is.
After answering, verify your response and mention if anything is missing.
"""

print("Cognitive Verifier Output:\n")
print(ask_ai(cognitive_verifier_prompt))
print("-" * 80)

# 4. Question Refinement Pattern
question_refinement_prompt = """
Refine the following question to make it clearer:
"What is AI?"
"""

print("Question Refinement Output:\n")
print(ask_ai(question_refinement_prompt))
print("-" * 80)

# 5. Provide New Information and Ask Questions
new_info_prompt = """
Explain Machine Learning.
Provide new related information.
Then ask two follow-up questions.
"""

print("Provide New Info & Ask Questions Output:\n")
print(ask_ai(new_info_prompt))
print("-" * 80)
```

Exercise 4
Chain of Thought, Tabular Format, Fill in the Blank, RGC, Zero-shot, One-shot, Few-shot Prompting

```python
print("\n=== EXERCISE 4 ===\n")

# 1. Chain of Thought
cot_prompt = """
Explain step by step how AI detects spam emails.
Provide a clear, structured explanation.
"""

print("Chain of Thought Output:\n")
print(ask_ai(cot_prompt))
print("-" * 80)

# 2. Tabular Format
tabular_prompt = """
Explain the difference between AI, ML, and Deep Learning in a table.
"""

print("Tabular Format Output:\n")
print(ask_ai(tabular_prompt))
print("-" * 80)

# 3. Fill in the Blank
fill_blank_prompt = """
Fill in the blank:
Artificial Intelligence is the ability of machines to ______ human intelligence.
"""

print("Fill in the Blank Output:\n")
print(ask_ai(fill_blank_prompt))
print("-" * 80)

# 4. RGC Prompting
rgc_prompt = """
Role: You are an AI mentor.
Goal: Explain prompt engineering.
Context: The learner is a beginner with no AI background.
"""

print("RGC Prompt Output:\n")
print(ask_ai(rgc_prompt))
print("-" * 80)

# 5. Zero-Shot Prompting
zero_shot_prompt = """
Explain Natural Language Processing.
"""

print("Zero-Shot Output:\n")
print(ask_ai(zero_shot_prompt))
print("-" * 80)

# 6. One-Shot Prompting
one_shot_prompt = """
Example:
Input: What is AI?
Output: AI is the simulation of human intelligence by machines.

Now explain:
What is Machine Learning?
"""

print("One-Shot Output:\n")
print(ask_ai(one_shot_prompt))
print("-" * 80)

# 7. Few-Shot Prompting
few_shot_prompt = """
Input: AI
Output: Machines that simulate human intelligence.

Input: ML
Output: Systems that learn from data.

Input: Deep Learning
Output:
"""

print("Few-Shot Output:\n")
print(ask_ai(few_shot_prompt))
print("-" * 80)
```
