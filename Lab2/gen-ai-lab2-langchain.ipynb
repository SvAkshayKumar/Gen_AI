```python
from kaggle_secrets import UserSecretsClient
from langchain_groq import ChatGroq
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain

# Step 1: Load Groq API Key from Kaggle Secrets
user_secrets = UserSecretsClient()
groq_api_key = user_secrets.get_secret("GROQ_API_KEY")

# Step 2: Initialize Groq Language Model
llm = ChatGroq(
    groq_api_key=groq_api_key,
    model_name="llama-3.1-8b-instant",
    temperature=0.7
)

# Step 3: Create Prompt Template
prompt = PromptTemplate(
    input_variables=["question"],
    template="Answer the following question clearly and concisely:\nQuestion: {question}"
)

# Step 4: Create LangChain LLM Chain
chain = LLMChain(
    llm=llm,
    prompt=prompt
)

# Step 5: Run the application
print("=== LangChain Q&A Application using Groq (Kaggle) ===")
question = input("Enter your question: ")

if question.strip() == "":
    print("Please enter a valid question.")
else:
    response = chain.run(question=question)
    print("\nAnswer:")
    print(response)
```
